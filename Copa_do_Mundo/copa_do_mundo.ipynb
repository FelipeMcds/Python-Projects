## Problema Copas Do Mundo

Considere um arquivo CSV com os seguintes dados de todas as copas do mundo até o ano de 2014. As informações do arquivo são:

* Year - Ano da copa
* Country - País sede
* Winner - Campeão
* Runners-Up - Vice-Campeão
* Third - Terceiro lugar
* Fourth - Quarto Lugar
* GoalsScored - Total de Gols
* QualifiedTeams - Quantidade de Times
* MatchesPlayed - Quantidade de Jogos
* Attendance - Público Total

Efetue a leitura desse arquivo, sem a utilização de módulos externos (Pandas), e crie um arquivo de saída com as seguintes informações (uma informação por linha):

* Soma de público das copas com anos final 0 (1930, 1950, etc)
* Quantidade total de gols entre as copas de 1954 e 1990, inclusive
* Média de público
* Média de gols por partida
* Quantidade de vezes em que o país sede foi campeão
* Quantidade de vezes em que o time do Brasil ficou entre uma das 4 primeiras posições
* Ano das edições em que o time da França finalizou em terceiro lugar
* Quantidade de vitórias por país, ordenada de forma crescente por número de títulos

```python
import csv
import re

somapublicofim0=0
goals54_90=0
publicototal=0
partidastotal=0
goalstotal=0
paisedechampion=0
brasil4lugares=0
franca3lugar=[]
qtdecopas=0
titulos=[]
campeoes={}


with open('csv/world_cups.csv','r',encoding='utf8') as cup:
    file = csv.reader(cup)
    next(file)#pula o cabeçalho
    for coluna in file:
        qtdecopas+=1  #Verifica o Número de copas ocorridas
        publicototal+=int(coluna[9].replace('.','')) #Soma a quantidade de público de todas as copas
        partidastotal+=int(coluna[8]) #Soma a quantidade de partidas de todas as copas
        goalstotal+=int(coluna[6]) #Soma a quantidade de gols de todas as copas

        if(coluna[0].endswith('0')): #verifica se o ano da copa é final 0, está em formato de string
            somapublicofim0+=int(coluna[9].replace('.',''))

        ano=int(coluna[0]) #Converto o ano que está em string para uma variável tipo int
        if(ano >=1954 and ano<=1990): #verifico se o ano está entre 1954 e 1990
            goals54_90+=int(coluna[6]) #somo só a quantidade de gols das copas que estão no intervalo anterior

        if(coluna[1]==coluna[2]): #verifica se a coluna de país sede tem o mesmo conteúdo da de campeão
            paisedechampion+=1

        if(coluna[2]=='Brazil' or coluna[3]=='Brazil' or coluna[4]=='Brazil' or coluna[5]=='Brazil'): #verifico quantas vezes o Brasil ficou nas 4 primeiras posições
            brasil4lugares+=1

        if(coluna[4]=='France'):#verifico quantas vezes a França ficou em Terceiro Lugar
            franca3lugar.append(coluna[0])

        if campeoes.get(coluna[2]) is None: #verifico se a chave que eu quero adicionar não existe no dicionário
            campeoes[coluna[2]]=1           #se el não existe eu adiciono e coloco o seu valor como 1
        else:
            campeoes[coluna[2]]+=1          #se ela já existe eu adiciono 1 título

    cup.close()

f= open('world_cupresp.txt','w',encoding='utf-8')

f.write(f'A Soma de Público das Copas com Final 0 foi de: {somapublicofim0}\n')
f.write(f'A quantidade de gols das copas de 1954 a 1990 foi de: {goals54_90}\n')
f.write(f'A média de público nas copas foi de: {(publicototal/qtdecopas):.2f}\n')
f.write(f'A média de gols das copas foi de: {(goalstotal/partidastotal):.2f}\n')
f.write(f'A quantidade de vezes que o país sede foi campeão da copa foi de: {paisedechampion} Vezes\n')
f.write(f'A quantidade de vezes que o Brasil ficou nas 4 primeiras posições foi de: {brasil4lugares} Vezes\n')
f.write(f'Os anos que a França ficaram em terceiro lugar foram os de: {franca3lugar}\n')
for i in sorted(campeoes,key = campeoes.get):
    f.write(f'{i}:{campeoes[i]}\n')
f.close()
```
